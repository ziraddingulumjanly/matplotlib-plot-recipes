```python
"""
Six mathematical functions displayed as a clean 3x2 Matplotlib subplot figure.

"""

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 600
plt.rcParams["savefig.dpi"] = 600
plt.rcParams["mathtext.fontset"] = "cm"
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.size"] = 18  


niceblue = "#4d7da7"
nicered = "#e15656"
nicegreen = "#6dc350"
nicegray = "#6e6e6e"
nicemagenta = "#c55aaa"
niceorange = "#d88c45"

save_plot = True

x = np.linspace(-5, 5, 1000)

plt.figure(figsize=(12, 11))

# Quadratic function
plt.subplot(3, 2, 1)
plt.plot(x, x**2, color=niceblue, linewidth=2.5)
plt.title(r"$y = x^2$", fontsize=14, pad=10)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(True, linestyle="--", alpha=0.4)

# Cubic function
plt.subplot(3, 2, 2)
plt.plot(x, x**3, color=nicered, linewidth=2.5)
plt.title(r"$y = x^3$", fontsize=14, pad=10)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(True, linestyle="--", alpha=0.4)

# Sine function
plt.subplot(3, 2, 3)
plt.plot(x, np.sin(x), color=nicegreen, linewidth=2.5)
plt.title(r"$y = \sin(x)$", fontsize=14, pad=10)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(True, linestyle="--", alpha=0.4)

# Cosine function
plt.subplot(3, 2, 4)
plt.plot(x, np.cos(x), color=nicemagenta, linewidth=2.5)
plt.title(r"$y = \cos(x)$", fontsize=14, pad=10)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(True, linestyle="--", alpha=0.4)

# Exponential function
plt.subplot(3, 2, 5)
plt.plot(x, np.exp(x / 2), color=niceorange, linewidth=2.5)
plt.title(r"$y = e^{x/2}$", fontsize=14, pad=10)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(True, linestyle="--", alpha=0.4)

# Hyperbolic tangent function
plt.subplot(3, 2, 6)
plt.plot(x, np.tanh(x), color=nicegray, linewidth=2.5)
plt.title(r"$y = \tanh(x)$", fontsize=14, pad=10)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.grid(True, linestyle="--", alpha=0.4)

plt.suptitle(
    "Six Mathematical Functions",
    fontsize=20,
    fontweight="light",
    y=0.98,
)

plt.tight_layout()

# Save the high-resolution plot before displaying it
if save_plot:
    plt.savefig(
        "six_mathematical_functions.pdf",
        dpi=600,
        bbox_inches="tight",
    )

plt.show()
```

```python
"""
Six mathematical functions plotted together in one clean Matplotlib figure.

You can uncomment some of the lines such as plt.grid, plt.axvline, plt.axhline and
loc="lower right" to "upper left in plt.legend, to see some different lookings. I mostly prefer no grid.

"""

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 600
plt.rcParams["savefig.dpi"] = 600
plt.rcParams["mathtext.fontset"] = "cm"
plt.rcParams["font.family"] = "serif"

niceblue = "#4d7da7"
nicered = "#e15656"
nicegreen = "#6dc350"
nicegray = "#6e6e6e"
nicemagenta = "#c55aaa"
niceorange = "#d88c45"

save_plot = True

# A smaller range keeps x³ from hiding the other functions
x = np.linspace(-2.5, 2.5, 1000)

# Mathematical functions
y1 = x**2
y2 = x**3
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.exp(x / 2)
y6 = np.tanh(x)

plt.figure(figsize=(10, 6))

plt.plot(x, y1, color=niceblue, linewidth=2.5, label=r"$y=x^2$")
plt.plot(x, y2, color=nicered, linewidth=2.5, label=r"$y=x^3$")
plt.plot(x, y3, color=nicegreen, linewidth=2.5, label=r"$y=\sin(x)$")
plt.plot(x, y4, color=nicemagenta, linewidth=2.5, label=r"$y=\cos(x)$")
plt.plot(x, y5, color=niceorange, linewidth=2.5, label=r"$y=e^{x/2}$")
plt.plot(x, y6, color=nicegray, linewidth=2.5, label=r"$y=\tanh(x)$")

# plt.axhline(0, color="black", linewidth=0.8, alpha=0.5)
# plt.axvline(0, color="black", linewidth=0.8, alpha=0.5)

plt.title("Six Mathematical Functions", fontsize=18, fontweight="bold", pad=12)
plt.xlabel(r"$x$", fontsize=13)
plt.ylabel(r"$f(x)$", fontsize=13)

plt.grid(False)
# plt.grid(True, linestyle="--", alpha=0.4)


plt.legend(
    loc="lower right",
    frameon=True,
    facecolor="white",
    edgecolor="#e0e0e0",
    fontsize=11,
)

plt.xlim(-2.5, 2.5)
plt.tight_layout()

# Save before displaying
if save_plot:
    plt.savefig(
        "six_mathematical_functions.pdf",
        bbox_inches="tight",
    )

plt.show()
```
