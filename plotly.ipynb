# Matplotlib — Your Plotting Materials

A focused Matplotlib project for creating clean, publication-style mathematical and data visualizations in Python. It uses your Computer Modern math styling and MATLAB-inspired palette.

## Shared visual style

Run this cell once at the beginning. The remaining cells use the same palette and font settings.

```python
"""Imports, high-resolution output, and the shared color palette."""

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 600
plt.rcParams["savefig.dpi"] = 600
plt.rcParams["mathtext.fontset"] = "cm"
plt.rcParams["font.family"] = "serif"

niceblue = "#4d7da7"
nicered = "#e15656"
nicegreen = "#6dc350"
nicegray = "#6e6e6e"
nicemagenta = "#c55aaa"
```

## Bar chart

Use a bar chart when the x-values represent separate categories or positions.

```python
"""A single bar chart with labels, gridlines, and a legend."""

x_values = [1, 2, 4, 5, 8]
y_values = [3, 5, 7, 9, 10]

plt.figure(figsize=(6, 3.5))
plt.bar(x_values, y_values, color=niceblue, label="Relationship")

plt.xlabel("Description of x-axis")
plt.ylabel("Description of y-axis")
plt.title("Bar Chart")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

## Five mathematical functions — explicit plot controls

This version sets the line width directly inside each `plt.plot()` call.

```python
"""Five functions using explicit line widths, math labels, and the shared palette."""

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 600
plt.rcParams["savefig.dpi"] = 600
plt.rcParams["mathtext.fontset"] = "cm"
plt.rcParams["font.family"] = "serif"

niceblue = "#4d7da7"
nicered = "#e15656"
nicegreen = "#6dc350"
nicegray = "#6e6e6e"
nicemagenta = "#c55aaa"

x = np.linspace(-2.5, 2.5, 1000)

y1 = np.exp(0.5 * x - 1) * (x + 1) ** 2
y2 = np.sin(x) * np.cos(x)
y3 = np.tanh(x)
y4 = x**2 - 2
y5 = np.cos(2 * x)

plt.figure(figsize=(8, 5))

plt.plot(x, y1, color=niceblue, linewidth=2.5, label=r"$e^{0.5x-1}(x+1)^2$")
plt.plot(x, y2, color=nicered, linewidth=2.5, label=r"$\sin(x)\cos(x)$")
plt.plot(x, y3, color=nicegreen, linewidth=2.5, label=r"$\tanh(x)$")
plt.plot(x, y4, color=nicegray, linewidth=2.5, label=r"$x^2 - 2$")
plt.plot(x, y5, color=nicemagenta, linewidth=2.5, label=r"$\cos(2x)$")

plt.title(r"Comparison of Five Mathematical Functions", fontsize=14, pad=12)
plt.xlabel(r"$x$", fontsize=12)
plt.ylabel(r"$f(x)$", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(
    loc="upper left",
    frameon=True,
    facecolor="w",
    edgecolor="#e0e0e0",
    fontsize=10,
)

plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 3.5)
plt.tight_layout()

plt.show()
```

## Five mathematical functions — global size controls

This version puts shared text and line sizes at the top, so changing a single value updates the whole chart.

```python
"""Five functions with centralized typography and line-width settings."""

import numpy as np
import matplotlib.pyplot as plt

base_fontsize = 18
linewidth = 1.5

niceblue = "#4d7da7"
nicered = "#e15656"
nicegreen = "#6dc350"
nicegray = "#6e6e6e"
nicemagenta = "#c55aaa"

plt.rcParams["figure.dpi"] = 600
plt.rcParams["savefig.dpi"] = 600
plt.rcParams["mathtext.fontset"] = "cm"
plt.rcParams["font.family"] = "serif"

plt.rcParams["lines.linewidth"] = linewidth
plt.rcParams["axes.labelsize"] = base_fontsize
plt.rcParams["axes.titlesize"] = base_fontsize
plt.rcParams["xtick.labelsize"] = base_fontsize - 3
plt.rcParams["ytick.labelsize"] = base_fontsize - 3
plt.rcParams["legend.fontsize"] = base_fontsize - 4

x = np.linspace(-2.5, 2.5, 1000)

y1 = np.exp(0.5 * x - 1) * (x + 1) ** 2
y2 = np.sin(x) * np.cos(x)
y3 = np.tanh(x)
y4 = x**2 - 2
y5 = np.cos(2 * x)

plt.figure(figsize=(9, 6))

plt.plot(x, y1, color=niceblue, label=r"$e^{0.5x-1}(x+1)^2$")
plt.plot(x, y2, color=nicered, label=r"$\sin(x)\cos(x)$")
plt.plot(x, y3, color=nicegreen, label=r"$\tanh(x)$")
plt.plot(x, y4, color=nicegray, label=r"$x^2 - 2$")
plt.plot(x, y5, color=nicemagenta, label=r"$\cos(2x)$")

plt.title(r"Comparison of Five Mathematical Functions", pad=15)
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")

plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(loc="upper left", frameon=True, facecolor="w", edgecolor="#e0e0e0")

plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 3.5)
plt.tight_layout()

plt.show()
```

## Multiple plots in a 2 × 2 grid

Use a subplot grid when each function needs its own panel.

```python
"""Four mathematical functions in separate panels."""

x = np.linspace(-3, 3, 1000)

y_1 = np.exp(0.5 * x - 1) * (x + 1) ** 2
y_2 = np.sin(x) * np.cos(x)
y_3 = np.tanh(x)
y_4 = x**2 - 2

fig, axes = plt.subplots(2, 2, figsize=(10, 7), sharex=True)

axes[0, 0].plot(x, y_1, color=niceblue)
axes[0, 0].set_title(r"$e^{0.5x-1}(x+1)^2$")
axes[0, 0].set_ylabel(r"$f(x)$")

axes[0, 1].plot(x, y_2, color=nicered)
axes[0, 1].set_title(r"$\sin(x)\cos(x)$")

axes[1, 0].plot(x, y_3, color=nicegreen)
axes[1, 0].set_title(r"$\tanh(x)$")
axes[1, 0].set_xlabel(r"$x$")
axes[1, 0].set_ylabel(r"$f(x)$")

axes[1, 1].plot(x, y_4, color=nicegray)
axes[1, 1].set_title(r"$x^2 - 2$")
axes[1, 1].set_xlabel(r"$x$")

for axis in axes.ravel():
    axis.grid(True, linestyle="--", alpha=0.5)

fig.suptitle("Functions in Separate Panels", y=1.02)
fig.tight_layout()

plt.show()
```
